```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("train.csv")
df.head()
```

```python
# matrix corelation 
df_numerik = df.select_dtypes(include=['int64', 'float64'])
matriks_korelasi = df_numerik.corr(method='pearson')
print(matriks_korelasi)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(matriks_korelasi, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap corelation pearson between variables')
plt.show()
```
